```python
import csv
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns; sns.set()  # for plot styling
import numpy as np
import platform

from matplotlib import font_manager, rc
plt.rcParams['axes.unicode_minus'] = False

if platform.system() == 'Darwin':
    rc('font', family='AppleGothic')
elif platform.system() == 'Windows':
    path = "c:/Windows/Fonts/malgun.ttf"
    font_name = font_manager.FontProperties(fname=path).get_name()
    rc('font', family=font_name)
else:
    print('Unknown system... sorry~~~~') 
```

```python
data=pd.read_csv('clustering.csv', encoding='utf-8-sig')
data=data.groupby('행정구').mean()
del data['생존율']
del data['생존율_3']
data
```

```python
plt.figure()
data['매출액'].sort_values().plot(kind='barh', grid=True, figsize=(10, 10))
plt.title('매출액 순위')
plt.show()
```

```python
plt.figure()
data['건수'].sort_values().plot(kind='barh', grid=True, figsize=(10, 10))
plt.title('건수 순위')
plt.show()
```

```python
plt.figure(figsize=(14,10))
plt.scatter(data['임대료'], data['매출액'], s=50)

for n in range(25):
    plt.text(data['임대료'][n]*1.02, data['매출액'][n]*0.99, 
             data.index[n], fontsize=15)

plt.xlabel('임대료')
plt.ylabel('매출액')
plt.axis([80000, 220000, 900, 1700])
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(14,10))
plt.scatter(data['임대료'], data['건수'], s=50)

for n in range(25):
    plt.text(data['임대료'][n]*1.02, data['건수'][n]*0.99, 
             data.index[n], fontsize=15)

plt.xlabel('임대료')
plt.ylabel('건수')
plt.grid()
plt.show()
```

```python
plt.figure()
np.round(data['음식점 수']/data['인구 수'], 4).sort_values(ascending=True).plot(kind='barh', grid=True, figsize=(10, 10))
plt.title('인구수 대비 음식점 수')
plt.show()
```
